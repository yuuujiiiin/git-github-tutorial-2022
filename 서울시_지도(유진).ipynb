```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
seoul_file = "/content/drive/MyDrive/Colab Notebooks/TL_KODIS_BAS.shp"
seoul = gpd.read_file(seoul_file, encoding='euckr')
seoul.tail(10)
```

```python
seoul2 = seoul[['SIG_CD', 'SIG_KOR_NM', 'geometry']]
seoul2
```

```python
seoul3 = seoul2.set_index("SIG_KOR_NM").sort_index()
seoul3 
```

```python
ax = seoul.plot(color='white', edgecolor='black', figsize=(15, 15))
ax.set_title("서울특별시 기초구역도")
ax.set_axis_off()
plt.show()
```

```python
seoul.geometry = seoul.buffer(0.001)
seoul = seoul.dissolve(by='SIG_CD')
ax = seoul.plot(figsize=(15, 15), column="SIG_KOR_NM", categorical=True,
                cmap="tab20b", edgecolor="k", legend=True, legend_kwds={'loc': 3})
ax.set_title("구 별로 묶은 서울의 기초 구역도")
ax.set_axis_off()
plt.show()
```
